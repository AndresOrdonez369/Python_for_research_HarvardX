```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


birddata=pd.read_csv("bird_tracking.csv")
bird_names=np.unique(birddata.bird_name)
plt.figure(figsize=(7,7))
for bird_name in bird_names:
    ix=birddata.bird_name==bird_name
    x,y=birddata.longitude[ix],birddata.latitude[ix]
    plt.plot(x,y,".", label=bird_name)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.savefig("3birds.pdf")
```

```python
speed=birddata.speed_2d[birddata.bird_name=="Eric"]
ind=np.isnan(speed)

plt.hist(speed[~ind], bins=np.linspace(0,30,20), density=True)
plt.xlabel("2D speed (m/s)")
plt.ylabel("Frequency")

birddata.speed_2d.plot(kind="hist",range=[0,30])
plt.xlabel("2D speed")
plt.savefig("pd_hist.pdf")
```

```python
birddata.columns
birddata.date_time[0:3]

import datetime
time_1=datetime.datetime.today()
time_2=datetime.datetime.today()
time_3=time_2 - time_1

date_str=birddata.date_time[0]
type(date_str)
date_str[:-3]
datetime.datetime.strptime(date_str[:-3],"%Y-%m-%d %H:%M:%S")

timestamps=[]
for k in range(len(birddata)):
    timestamps.append(datetime.datetime.strptime(birddata.date_time.iloc[k][:-3],"%Y-%m-%d %H:%M:%S"))
timestamps[0:3]
birddata["timestamp"]=pd.Series(timestamps, index=birddata.index)

times=birddata.timestamp[birddata.bird_name=="Eric"]
elapsed_times=[time-times[0] for time in times]
elapsed_times[1000]/datetime.timedelta(days=1)
elapsed_times[1000]/datetime.timedelta(hours=1)
plt.figure(figsize=(7,7))
plt.plot(np.array(elapsed_times)/datetime.timedelta(days=1))
plt.xlabel("Observations")
plt.ylabel("Elapsed times(days)")

```

```python
data=birddata[birddata.bird_name=="Eric"]
times=data.timestamp
elapsed_times=[time-times[0] for time in times]
elapsed_days=np.array(elapsed_times)/datetime.timedelta(days=1)
next_day=1
inds=[]
daily_mean_speed=[]
for (i,t) in enumerate(elapsed_days):
    if t < next_day:
        inds.append(i)
    else:
        daily_mean_speed.append(np.mean(data.speed_2d[inds]))
        next_day+=1
        inds=[]

plt.figure(figsize=(8,6))
plt.plot(daily_mean_speed)
plt.xlabel("Day")
plt.ylabel("Main speed (m/s)")
```
