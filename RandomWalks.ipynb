```
import numpy as np
import matplotlib.pyplot as plt
X_0=np.array([[0],[0]])
delta_X=np.random.normal(0,1,(2,100))
X=np.concatenate((X_0,np.cumsum(delta_X,axis=1)),axis=1)

plt.figure()
plt.subplot(121)
plt.plot(X[0],X[1],"ro-",label="Con origen")
plt.legend(loc="best")
plt.subplot(122)
plt.plot(delta_X[0],delta_X[1],"bo-",label="Solo los datos")
plt.legend(loc="best")
plt.savefig("Random Walkfinal.pdf")
```
